```python
Dim = 6
mean = np.random.uniform(-0.7,3.0, Dim)
cov_sign = np.sign(np.random.uniform(-1.,1., (Dim,Dim) ))
cov_sign = cov_sign * cov_sign.T

covm = np.random.uniform(0.8,1.0, (Dim,Dim))
covm = covm * covm.T * cov_sign

np.fill_diagonal(covm, np.random.uniform(1.0, 2.0, Dim))
```

```python
D = pd.DataFrame(np.random.multivariate_normal(mean, covm, 1000),
                 columns = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")[:Dim]
                )
```

```python
g = sns.PairGrid(D, diag_sharey=False)
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_upper(plt.scatter, s=5)
g.map_diag(sns.kdeplot, lw=3)
```
